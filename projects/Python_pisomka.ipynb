4. Definujte funkciu, ktorá vráti menšie z 2 čísel ak su obe párne a väčšie ak čísla párne niesu.

```python
# Kod
def c(cislo1, cislo2):
    if cislo1%2==0 and cislo2%2==0:
        if cislo1 < cislo2:
            print(cislo1)
        else:
            print(cislo2)
            
    else:
        if cislo1 > cislo2:
            print(cislo1)
        else:
            print(cislo2)

c(4,6)
```

5. Definujte funkciu, ktorá vráti True ak je súčet 2 zadaných čísel 30 alebo je jedno číslo 30, inak vráti False.

```python
def makes_thirty(a,b):

    if a==30 or b==30 or a+b==30:
        return True
    else:
        return False

print(makes_thirty(10,20))
print(makes_thirty(30,5))
print(makes_thirty(35,5))
```
